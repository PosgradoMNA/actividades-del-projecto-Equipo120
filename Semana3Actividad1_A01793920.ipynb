**3.1 Base de datos y Almacén de datos**

Este primer concepto se define como el programa de software que almacena información relacionada con una actividad o propósito específico, por otro lado, el almacén de datos (también llamado data warehouse) es una base de datos que contiene información de muchas fuentes diferentes y las transforma en una especie de esquema que nos permite consultar estos diferentes datos de manera simultánea y nos permite hacer insights, a través del análisis de datos. (Kane, 2017).

ETL: extraer, transformar y cargar (almacenamiento de datos de manera convencional)

ELT: pasar de una base de datos monolítica creada en Oracle o MySQL, a una de estas bases de datos distribuidas más modernas creadas sobre Hadoop, Hive, Spark o MapeReduce.

Una base de datos es una colección de tablas relacionadas, y un esquema de base de datos almacena información sobre las tablas (y otros objetos de la base de datos), así como las relaciones entre ellos, definiendo la estructura de la base de datos, y los almacenes de datos a menudo contienen datos de múltiples fuentes de datos subyacentes y se diseñan utilizando técnicas de modelado dimensional. (Teate, 2021)

**3.2 Uniformidad y Pre-procesamiento de datos**

El preprocesamiento de datos consiste en transformar datos sin procesar en un formato comprensible.

Las tareas en el preprocesamiento de datos son: (i) Limpieza de datos: rellenar los valores faltantes, suavizar los datos ruidosos, identificar o eliminar los valores atípicos y resolver las incoherencias, (ii) Transformación de datos: Normalización y (iii) Reducción de datos: reducir el volumen pero producir el mismo análisis o uno similar.

La limpieza de datos se define como el proceso de identificar y corregir entradas erróneas de un marco de datos, tabla o base de datos y se refiere a detectar partes incompletas o irrelevantes de los datos y luego reemplazar, modificar o eliminar esos datos. (Mckinney, 2017)

**Manejo de datos faltantes**

Se definen como el valor de datos que no se almacena para una variable en la observación de interés, pueden producir resultados sesgados, lo que lleva a conclusiones no válidas. 

**Tipos de datos faltantes:**

•	Falta completamente al azar: si la probabilidad de que falten los datos no está relacionada ni con el valor ni con el conjunto de respuestas observadas.
•	Falta al azar: si la probabilidad de que falten las respuestas depende del conjunto de respuestas observadas
•	Falta no al azar: cuando los datos faltantes no siguen las características del MCAR o MAR.

**Técnicas para manejar los datos faltantes**

•	Eliminación por lista: eliminar aquellas instancias con los datos que faltan y analizar los datos restantes
•	Eliminación por parejas: solo cuando falta el punto de datos necesario para probar una suposición particular
•	Sustitución media: cuando el valor medio de una variable se usa en lugar del valor de datos faltantes para esa misma variable
•	Imputación de regresión: proceso de reemplazar los datos faltantes con valores estimados
•	Imputación Múltiple: en lugar de sustituir un solo valor por cada dato faltante, los valores faltantes se reemplazan con un conjunto de valores razonables que contienen las variaciones de los valores correctos

(Grace-Martin, 2022)
