##**Metodología CRISP-DM**##

La metodología que se seleccionó es CRISP-DM, es un modelo de proceso que sirve como base para un proceso de ciencia de datos.

Como metodología, incluye descripciones de las fases normales de un proyecto, las tareas necesarias en cada fase y una explicación de las relaciones entre las tareas y como modelo de proceso, CRISP-DM ofrece un resumen del ciclo vital de minería de datos.

##**Pasos y Procesos de la Metodología**##

**1. Comprensión empresarial**

Comprensión de objetivos y requisitos del proyecto desde una perspectiva empresarial, generando un plan preliminar.

**2. Comprensión de datos**

Recopilación y familiarización de datos,  identificación de y descubrir los primeros conocimientos sobre los datos. 

**3. Preparación de datos**

Cubre todas las actividades para construir el conjunto de datos final a partir de los datos sin procesar iniciales.

**4. Modelado**

Evalúa, selecciona y aplica las técnicas de modelado apropiadas. (com posible bucle de regreso)

**5. Evaluación**

Crear y elegir  modelos que parecen tener alta calidad en función de las funciones de pérdida que se seleccionaron, se prueban y se validan, el resultado final es la selección de los modelos campeones.

**6. Despliegue**

Implementar una representación de código del modelo en un sistema operativo, incluyendo mecanismos para puntuar o categorizar nuevos datos no vistos a medida que surgen. El modelo tratará los nuevos datos sin procesar de la misma manera que durante el desarrollo del modelo.
