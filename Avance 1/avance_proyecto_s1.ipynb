```python
# Import necessary libraries:
import os
import pandas as pd
import warnings
import webbrowser
```

```python
# Define the constants:
DATASET_URL = "https://www.kaggle.com/datasets/rikdifos/credit-card-approval-prediction/download?datasetVersionNumber=3"
DATASET_PATH = "../data/Credit Card Approval Prediction/"
```

```python
def showcaseDataset(filename: str) -> pd.DataFrame:
    """
    Simple showcasing function to avoid program crashes when the dataset 
    files are not found in the expected local route.
    """
    try:
        df = pd.read_csv(os.path.join(DATASET_PATH, filename))
        return df.head()
    except:
        print("File {} not found in the data path.".format(filename))
        warnings.warn("[FILE NOT FOUND] Opening the download URL fo the data. Please save the downloaded CSV files in the '{}' local folder.".format(DATASET_PATH))
        webbrowser.open(DATASET_URL)
    
```

```python
print("Showcasing the main dataset to use through the module:")

showcaseDataset("application_record.csv")
```

```python
print("Showcasing the complementary dataset to use through the module:")

showcaseDataset("credit_record.csv")
```
